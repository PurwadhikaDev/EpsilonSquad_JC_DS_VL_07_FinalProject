## Deteksi Outlier

```python
sns.boxplot(data=df[["windspeed", "hum","temp"]], orient="v")
```

Terlihat terdapat outlier di kolom windspeed dan hum. Namun tidak terlihat di kolom temperatur. Berikutnya kita akan deteksi data poin mana saja dalam kolom tersebut yang dikategorikan sebagai outlier


```python
#Membuat function untuk mendeteksi outlier
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}') 
```

Pertama, kita ingin mendeteksi outliers di kolom secara bergantian. Pertama, kita menganalisa kolom humidity. Setelahnya, beralih ke kolom

```python
#Import Library yang diperlukan

#Model & Pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import TransformedTargetRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV 
from sklearn.neighbors import KNeighborsRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.svm import SVR
from sklearn.linear_model import ElasticNet
from prophet import Prophet


#Transform
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import SplineTransformer #Akan kami gunakan dalam Feature Engineering
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import RobustScaler

#Validation
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV, TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

#Kami membuat variabel baru yang akan digunakan untuk modelling nantinya.
poly = PolynomialFeatures(degree=3,interaction_only=False, include_bias=False) #Untuk membuat polinomial features
onehot = OneHotEncoder() #Tranformasi Onehot untuk fitur kategorikal
alphas = np.logspace(-6, 6, 25) #Untuk RidgeCV
kolom_num = ['hum','windspeed','temp'] #Kolom numerikal
kolom_kat = ['season','weathersit','holiday','workingday'] #Kolom kategorikal

```

## Encoding

```python
#Encoding
def periodic_spline_transformer(period, n_splines=None, degree=3):
    if n_splines is None:
        n_splines = period
    n_knots = n_splines + 1  # periodic and include_bias is True
    return SplineTransformer(
        degree=degree,
        n_knots=n_knots,
        knots=np.linspace(0, period, n_knots).reshape(n_knots, 1),
        extrapolation="periodic",
        include_bias=True,
    )

##Kami menggunakan spline transformer karena metode tersebut dinilai paling baik untuk data kami yang sifatnya Time Series
#mesikipun begitu, statusny bisa saja berubah pada dataset yang berbeda. Maka dari itu direkomendasikan untuk
#menggunakan beberapa metode terlebih dahulu sebagai perbandingan.


cyclic_spline_transformer = ColumnTransformer(
    transformers=[
        ("kategorikal", onehot, kolom_kat),
        ("cyclic_month", periodic_spline_transformer(12, n_splines=6), ["mnth"]), #12 dari banyaknya bulan/Tahun
        ("cyclic_weekday", periodic_spline_transformer(7, n_splines=3), ["weekday"]), #7 dari jumlah hari/minggu
        ("cyclic_hour", periodic_spline_transformer(24, n_splines=12), ["hr"]), #24 dari jumlah jam/hari
        ("numerikal",poly, kolom_num)
    ],
    remainder=RobustScaler(), 
)
```

```python
# Define algoritma yang digunakan
lr = LinearRegression()
Ridge = RidgeCV(alphas=alphas)
svr = SVR(gamma='auto')
enet = ElasticNet()
knn = KNeighborsRegressor()
xgb = XGBRegressor(random_state=2022)
```

```python
# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)
log_rc = TransformedTargetRegressor(Ridge, func=np.log, inverse_func=np.exp)
log_svr = TransformedTargetRegressor(svr, func=np.log, inverse_func=np.exp)
log_enet = TransformedTargetRegressor(enet, func=np.log, inverse_func=np.exp)

#Kandidat model
models = [log_lr,log_rc, log_knn, log_xgb,log_svr,log_enet]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []
```

```python
# Mencari algoritma terbaik berdasarkan nilai RMSE
for model in models:
    
    crossval = ts_cv

    estimator = Pipeline([
        ('preprocessing', cyclic_spline_transformer),
        ('model', model)
    ])

    # RMSE Cross Validation
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    # print

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
        )

    # print

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

```python
# Evaluasi hasil dari 5 kandidat algoritma yang digunakan
pd.DataFrame({
    'Model': ['LinearR','Ridge','KNNR','XGBR','svr','enet'],
    'Mean_RMSE': nilai_mean_rmse,
    'Std_RMSE': nilai_std_rmse,
    'Mean_MAPE': nilai_mean_mape,
    'Std_MAPE': nilai_std_mape
})
```

XGboost dan SVR menjadi 2 kandidat teratas. Nilai RMSE dari xgboost lebih baik daripada svr. Sedangkan, svr lebih baik dalam hal standar deviasinya. Untuk MAPE, nilai standar deviasi svr dan nilainya lebih buruk. Selanjutnya, akan kami bandingkan dengan memprediksi pada Test Set

# Model Tuning

## Hyperparameter Tuning

Kami melakukan hyperparameter tuning dengan panduan seperti berikut :
Source : https://xgboost.readthedocs.io/en/latest/parameter.html#general-parameters

```python
# xgb
max_depth = list(np.arange(3, 11)) # Kedalaman pohon
learning_rate = list(np.arange(1, 4)/100) # Learning rate
n_estimators = list(np.arange(100, 1001)) # Jumlah pohon
subsample = list(np.arange(6, 11)/10) # Jumlah baris tiap pohon (% dari total baris train set)
gamma = list(np.arange(1, 6)/10) # Semakin besar nilainya, semakin konservatif/simpel modelnya
colsample_bytree = list(np.arange(5, 11)/10) # Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
reg_alpha = list(np.arange(5, 50)/100) # Semakin besar nilainya, semakin konservatif/simpel modelnya

min_child_weight = list(np.arange(1, 11)) # The larger min_child_weight is, the more conservative the algorithm will be
eta = [0.1]

# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha,
    'model__eta': eta,
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=2022)

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', cyclic_spline_transformer),
        ('model', xgb)
        ])

crossval = ts_cv

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 50,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_root_mean_squared_error', # Hanya bisa memilih salah stau metric untuk optimisasi
    random_state = 2022  
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

```python
# Cek skor dan params terbaik
print('XGBoost')
print('Best_score:', random_xgb.best_score_)
print('Best_params:', random_xgb.best_params_)
```

## Predict to Test Set with the Tuned Model

```python
# Model XGBoost
model = {'XGB': XGBRegressor(random_state=2022)}

# Define model terhadap estimator terbaik
tuned_model = random_xgb.best_estimator_

# Fitting model
tuned_model.fit(X_train, y_train)

# Predict test set
y_pred_tuned = tuned_model.predict(X_test)

# Simpan nilai metrics RMSE, MAE setelah tuning
rmse_xgb_tuned = np.sqrt(mean_squared_error(y_test, y_pred_tuned))
mape_xgb_tuned = mean_absolute_percentage_error(y_test, y_pred_tuned)

score_after_tuning = pd.DataFrame({'RMSE': rmse_xgb_tuned, 'MAPE': mape_xgb_tuned}, index=model.keys())
score_after_tuning
```
